```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/netflix_titles.csv")
```

```python
df.head()
```

```python
sns.countplot(x='type', data=df)
plt.title('Count of TV Shows and Movies')
plt.show()
```

```python
plt.figure(figsize=(15, 8))
top_countries = df['country'].value_counts().head(15)
sns.barplot(x=top_countries.values, y=top_countries.index, palette='viridis')
plt.title('Top 15 Countries with Most Content on Netflix')
plt.xlabel('Number of Shows/Movies')
plt.show()
```

```python
plt.figure(figsize=(15, 8))
top_countries = df['country'].value_counts().head(15)
sns.barplot(x=top_countries.values, y=top_countries.index, palette='viridis')
plt.title('Top 15 Countries with Most Content on Netflix')
plt.xlabel('Number of Shows/Movies')
plt.show()
```

```python
plt.pie(df['type'].value_counts(), labels=df['type'].value_counts().index, autopct='%1.1f%%')
plt.title('Distribution of TV Shows and Movies on Netflix')
plt.show()
```

```python
plt.figure(figsize=(15, 8))
top_directors = df['director'].value_counts().head(10)
sns.barplot(x=top_directors.values, y=top_directors.index, palette='muted')
plt.title('Top 10 Directors with Most Content on Netflix')
plt.xlabel('Number of Shows/Movies')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
rating_dist = df['rating'].value_counts().sort_index()
sns.barplot(x=rating_dist.index, y=rating_dist.values, palette='mako')
plt.title('Distribution of Ratings on Netflix')
plt.xlabel('Rating')
plt.ylabel('Number of Shows/Movies')
plt.show()
```

```python
plt.figure(figsize=(15, 8))
genre_dist = df['listed_in'].value_counts().head(15)
sns.barplot(x=genre_dist.values, y=genre_dist.index, palette='rocket')
plt.title('Top 15 Genres on Netflix')
plt.xlabel('Number of Shows/Movies')
plt.show()
```

```python

director_name = "Christopher Nolan"
movies_by_director = df[df['director'].str.contains(director_name, case=False, na=False)]

print(f"Movies directed by {director_name} on Netflix:")
print(movies_by_director[['show_id', 'title', 'type', 'release_year']])
```

```python
type_counts = df['type'].value_counts()

# Create a line plot
plt.figure(figsize=(8, 5))
sns.lineplot(x=type_counts.index, y=type_counts.values, marker='o', color='blue')
plt.title('Total Count of Movies and TV Shows on Netflix')
plt.xlabel('Content Type')
plt.ylabel('Total Count')
plt.show()
```
